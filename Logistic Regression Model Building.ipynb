```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.stats as stats
sns.set()
import warnings
warnings.filterwarnings("ignore")
import statsmodels.api as sm
```

```python
data = pd.read_csv("2.02. Binary predictors.csv")
```

```python
data.head()
```

```python
data_1 = data.copy()
data_1["Admitted"] = data["Admitted"].map({"Yes":1,"No":0})
data_1["Gender"] = data["Gender"].map({"Female":1,"Male":0})
```

```python
data_1.head()
```

```python
#declaring dependent and independent
```

```python
y = data_1["Admitted"]
x1 = data_1[["SAT","Gender"]]
```

```python
# building model
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results = reg_log.fit()
results
```

```python
results.summary()
```

```python
# Accuracy
```

```python
k = results.predict()
k.round(2)
```

```python
np.array(data_1["Admitted"])
```

```python
df =  pd.DataFrame(a)
df.columns = ["Predicted 0","Predicted 1"]
df.rename(index={0:"Actual 0",1:"Actual 1"},inplace=True)
df #Confusion matrix
```

```python
cm = np.array(df)
accuracy = (cm[0,0]+cm[1,1])/cm.sum()*100
accuracy
```

```python
def confusion_matrix(data,actual_values,model):
    
        pred_values = model.predict(data)
        bins=np.array([0,0.5,1])
        cm = np.histogram2d(actual_values, pred_values, bins=bins)[0]
        accuracy = (cm[0,0]+cm[1,1])/cm.sum()
        return cm, accuracy
```
